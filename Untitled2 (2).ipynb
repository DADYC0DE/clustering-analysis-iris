```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN

from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
iris = load_iris()
X = iris.data
feature_names = iris.feature_names

df = pd.DataFrame(X, columns=feature_names)
df.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
K_range = range(2, 10)

for k in K_range:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X_scaled)
    inertia.append(model.inertia_)

plt.figure()
plt.plot(K_range, inertia, marker='o')
plt.title("Elbow Method (Inertia vs K)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.grid()
plt.show()
```

```python
silhouette_scores = []

for k in K_range:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled, labels))

plt.figure()
plt.plot(K_range, silhouette_scores, marker='o')
plt.title("Silhouette Score vs K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Score")
plt.grid()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels_kmeans = kmeans.fit_predict(X_scaled)
```

```python
agg = AgglomerativeClustering(n_clusters=3)
labels_agg = agg.fit_predict(X_scaled)
```

```python
db = DBSCAN(eps=0.5, min_samples=5)
labels_db = db.fit_predict(X_scaled)
```

```python
plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels_kmeans)
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels_agg)
plt.title("Agglomerative Clustering")
plt.show()
```

```python
plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels_db)
plt.title("DBSCAN Clustering")
plt.show()
```

```python
linked = linkage(X_scaled, method='ward')

plt.figure(figsize=(10,5))
dendrogram(linked)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```

```python
def evaluate(name, X, labels):
    print(f"\n{name}")
    if len(set(labels)) > 1:
        print("Silhouette:", silhouette_score(X, labels))
        print("Calinski-Harabasz:", calinski_harabasz_score(X, labels))
        print("Davies-Bouldin:", davies_bouldin_score(X, labels))
    else:
        print("Invalid clustering")

evaluate("KMeans", X_scaled, labels_kmeans)
evaluate("Agglomerative", X_scaled, labels_agg)
evaluate("DBSCAN", X_scaled, labels_db)
```

```python
models = ["KMeans", "Agglomerative", "DBSCAN"]

sil_scores = [
    silhouette_score(X_scaled, labels_kmeans),
    silhouette_score(X_scaled, labels_agg),
    silhouette_score(X_scaled, labels_db)
]

plt.figure()
plt.bar(models, sil_scores)
plt.title("Model Comparison (Silhouette Score)")
plt.xlabel("Algorithms")
plt.ylabel("Score")
plt.show()
```

```python
# Try different K
for k in [2, 3, 4]:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X_scaled)
    print(f"K={k}, Silhouette:", silhouette_score(X_scaled, labels))
```
